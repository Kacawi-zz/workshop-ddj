### Functies en "Packages"

#### Functies
- Lossen een bepaald, goed gedefinieerd probleem op 
- Black-box principe
- Heel veel functies beschikbaar in packages 
- Je kan ook zelf functies schrijven!


```python
def my_mean(lst):
    sum = 0.0;
    for el in lst:
        sum += el;
    return sum / len(lst)

scores = [12, 17, 19, 9]
my_mean(scores)
```

- Gemakkelijke installatie met `pip`

Wat weten jullie al over "fake ads"?

Trackers zijn snippets van code die genstalleerd worden door websites zodat ze ads van ad networks kunnen tonen aan hun bezoekers en tegelijkertijd ook data kunnen verzamelen over die bezoekers.

Buzzfeed News heeft data verzameld om te proberen te onderzoeken welke trackers toegevoegd of verwijderd werden van ad networks sinds November van vorig jaar, toen er algemene consternatie was over "Fake news". In andere woorden, we willen het effect onderzoeken van de heisa op de aan-of afwezigheid van trackers op websites. 

We kunnen dit onderzoeken met de hulp van Python!

#### Verzamelen, Importeren & Exploreren van Data

Natuurlijk, om de bovenstaande vraag te beantwoorden, hebben we de data nodig. De data is in dit geval voor ons verzameld. Wij zullen focussen op het importeren en exploreren van de data in deze stap!

```python
# Import the pandas library
import pandas as pd

# Read in the observed trackers data
trackers = pd.read_csv("../data/observed-trackers.csv")

# Inspect the first 50 rows
trackers.head(50)
```

Om de situatie van voor en na november 2016 te vergelijken, kunnen we de `when` kolom gebruiken: deze lijst op welke ad trackers er aanwezig waren op sites vr november 2016, de maand waarin het eerst algemene bezorgdheid werd uitgedrukt over "fake news".

Laat ons dit in detail bekijken voor bepaalde domeinen:

```python
# Print the unqiue domains
print(trackers.domain.unique())

# Inspecting domains
trackers[trackers.domain =="kupr7.com"]

# Inspecting another domain
trackers[trackers.domain=="mckenziepost.com"]

# Inspecting yet another domain
trackers[trackers.domain=="20minutenews.com"]
```

We ondervinden twee zaken:
- Er zijn sites die geen geobserveerde trackers hebben in de "before" periode
- Er zijn sites die verdwenen zijn in de "after" periode

Deze domeinen moeten weg als we het effect willen bekijken van voor en na de heisa over "fake news".

#### "Preprocessing" van Data

Een kleine fast-forward: we hebben alle "dead domains" opgelijst in een lijst `DEAD_DOMAINS` en gaan nu ook op zoek naar domeinen die vergelijkbaar zijn met degene die we opgelijst hebben, maar die 
- Geobserveerde trackers hebben in de "before" periode;
- Sites die niet in de lijst van `DEAD_DOMAINS` zitten.

We slagen deze 'nieuwe' data op in `comparable_domains`.

```python
DEAD_DOMAINS = [
    "abcnews.com.co", 
    "alynews.com", 
    "areyouasleep.com", 
    "baltimoregazette.com", 
    "channel16news.com", 
    "channel17news.com", 
    "channel18news.com", 
    "christiantimesnewspaper.com", 
    "clancyreport.com", 
    "dailynews11.com", 
    "denverguardian.com", 
    "heaviermetal.net", 
    "kspm33.com", 
    "kupr7.com", 
    "ky6news.com", 
    "kypo6.com", 
    "mbynews.com", 
    "mckenziepost.com", 
    "msnbc.website",
    "newsbuzzdaily.com",
    "newsnow17.com", 
    "newswatch33.com", 
    "oreillypost.com", 
    "scrapetv.com", 
    "thebostontribune.com", 
    "thereporterz.com", 
    "wleb21.com", 
]
```

```python
comparable_domains = trackers[
    (trackers["when"] == "before-nov-2016") &
    ~trackers["domain"].isin(DEAD_DOMAINS)
]["domain"].unique()
```

```python
print(len(trackers["domain"].unique()))

print(len(comparable_domains))
```

```python
comparable_trackers = trackers[
    trackers["domain"].isin(comparable_domains)
]
```

Van de 104 domeinen die in de dataset zaten, zijn er dus 51 die eventueel vergeleken kunnen worden op voor-en na november 2016.

Laten we nu een matrix maken dat ons een overzicht geeft van welke trackers er op welke websites gevonden zijn:

```python
tracker_matrix = comparable_trackers.groupby([
    "domain",
    "when",
    "tracker"
]).size().unstack().unstack() > 0
tracker_matrix.head()
```

Met dit resultaat in handen kunnen we gaan kijken naar specifieke ad networks, zoals Google Adsense:

```python
tracker_matrix["Google Adsense"]
```

We zien dat, bijvoorbeeld, Google AdSense tracker verwijderd werd door 10 websites tussen november 2016 en maart 2017. Dit uit een totaal van 51 sites, waarvan de huidge en de historische tracker data beschikbaar voor was.  

Wat voor soort visualisatie zou hier op zijn plaats zijn?

```python
tracker_counts = tracker_matrix.sum().unstack()

tracker_nov = pd.DataFrame(data=tracker_counts['before-nov-2016'], index=tracker_counts.index.values).sort_values(by=['before-nov-2016'], ascending=False)
tracker_after = pd.DataFrame(data=tracker_counts['march-2017'], index=tracker_counts.index.values).sort_values(by=['march-2017'], ascending=False)
```

```python
tracker_nov_small = tracker_nov.head(10).sort_values(by=['before-nov-2016'], ascending=True)
tracker_after_small = tracker_after.head(10).sort_values(by=['march-2017'], ascending=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt


tracker_nov_small.plot.barh()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt


tracker_after_small.plot.barh()
```

We slagen dit resultaat op in een CSV file:

```python
pd.DataFrame(
    tracker_matrix.values,
    index=tracker_matrix.index,
    columns=list(map("|".join, tracker_matrix.columns.values))
).to_csv("../tracker-matrix.csv")
```
